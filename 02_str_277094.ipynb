```python
import numpy as np
from matplotlib import pyplot as plt
import sympy
import random

x=sympy.Symbol("x")

x1=random.uniform(0,2)
x2=random.uniform(2,4)
x3=random.uniform(4,6)
x4=random.uniform(6,8)

x_1=round(x1,3)
x_2=round(x2,3)
x_3=round(x3,3)
x_4=round(x4,3)

x=np.linspace(0,8,100)

y=(x-x1)*(x-x2)*(x-x3)*(x-x4)


fig=plt.figure()
ax=fig.add_subplot()

ax.set_title('Losowe pierwiastki: '+str(x_1)+ ' ; '  +str(x_2)+ ' ; ' +str(x_3)+ ' ; ' +str(x_4)+'')

dydx=np.gradient(y,x)

ekstremum_x=x[1:][dydx[1:]*dydx[:-1]<0]
ekstremum_y=y[1:][dydx[1:]*dydx[:-1]<0]
ax.scatter([ekstremum_x[0],ekstremum_x[1],ekstremum_x[2]],[ekstremum_y[0],ekstremum_y[1],ekstremum_y[2]],color="black",zorder=3)
ax.set_xticks([ekstremum_x[0],ekstremum_x[1],ekstremum_x[2]])
ax.set_ylim(-70,70)

def wartosci_funkcji(lista_x,x1,x2,x3,x4):    
    lista_y=[]
    for i in lista_x:
        liczba=(i-x1)*(i-x2)*(i-x3)*(i-x4)
        lista_y.append(liczba)
    return lista_y

lista_x1=np.linspace(0,ekstremum_x[0],100)
lista_x2=np.linspace(ekstremum_x[0],ekstremum_x[1],100)
lista_x3=np.linspace(ekstremum_x[1],ekstremum_x[2],100)
lista_x4=np.linspace(ekstremum_x[2],8,100)

ax.plot(lista_x1,wartosci_funkcji(lista_x1,x1,x2,x3,x4),color="orange")
ax.plot(lista_x2,wartosci_funkcji(lista_x2,x1,x2,x3,x4),color="green")
ax.plot(lista_x3,wartosci_funkcji(lista_x3,x1,x2,x3,x4),color="orange")
ax.plot(lista_x4,wartosci_funkcji(lista_x4,x1,x2,x3,x4),color="green")


v1=round(abs(ekstremum_y[0])+abs(ekstremum_y[1]),3)
v2=round(abs(ekstremum_y[1])+abs(ekstremum_y[2]),3)

ax.annotate(str(v1),xy=((ekstremum_x[0]+ekstremum_x[1])/2,(ekstremum_y[0]+ekstremum_y[1])/2),xytext=((((ekstremum_x[0]+ekstremum_x[1])/2)-0.5),((ekstremum_y[0]+ekstremum_y[1])/2)+30),arrowprops=dict(shrink=0.05,facecolor='blue'))
ax.annotate(str(v2),xy=((ekstremum_x[1]+ekstremum_x[2])/2,(ekstremum_y[1]+ekstremum_y[2])/2),xytext=((((ekstremum_x[1]+ekstremum_x[2])/2)-0.5),((ekstremum_y[1]+ekstremum_y[2])/2)+30),arrowprops=dict(shrink=0.05,facecolor='red'))


ax.spines['right'].set_color('k')
ax.spines['top'].set_color('k')
ax.spines['bottom'].set_color('k')
ax.axhline(0,color="black",linewidth=0.75)
plt.legend(["Funkcja malejąca","Funkcja rosnąca"],loc=0)
plt.savefig("02_str_277094.pdf")
```
