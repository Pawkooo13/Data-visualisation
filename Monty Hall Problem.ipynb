```python
import numpy as np
from numpy import random as rd
from matplotlib import pyplot as plt

n = 1000
attempts = 100

list1 = []
list2 = []
list3 = []
list4 = []

doors = ['g','c','g']

for i in range(attempts):
    countswitch = 0
    countnoswitch = 0
    for j in range(n):
        rd.shuffle(doors) 
        #don't have guarantee that every posibilities is checked so number of n must be high
        #some variants can be repeated
        choose = rd.choice(doors)
        #the host reveal one goat and aks about switch
        if(choose == 'c'): #stay to door which was choosen
            countnoswitch += 1
        else:
            countswitch += 1
        if(len(list1) < n):
            list1.append(countnoswitch)
            list2.append(countswitch)
    list3.append(countnoswitch/n)
    list4.append(countswitch/n)
    
print('Average stay: ' , np.mean(list3))
print('Average switch: ' , np.mean(list4))
print('Max switch: ', np.max(list4))
print('Min switch: ', np.min(list4))
print('Max stay: ', np.max(list3))
print('Min stay: ', np.min(list3))

fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax.plot(list1)
ax.plot(list2)
ax.set_title(str(n) + " variants")

ax2.scatter(np.arange(0,attempts,1), list3, color='blue')
ax2.scatter(np.arange(0,attempts,1), list4, color='orange')
ax2.plot(np.arange(0,attempts,1), np.linspace(np.mean(list3),np.mean(list3),attempts), color='red')
ax2.plot(np.arange(0,attempts,1), np.linspace(np.mean(list4),np.mean(list4),attempts), color='red')
ax2.set_ylim([0,1])
ax2.set_title("Averages of " + str(attempts) + " attempts")
```
