```python
import numpy as np
from matplotlib import pyplot as plt
import sympy

x=sympy.Symbol("x")

funkcja=x**3+x**2-5*x

def wartosc_funkcji(x):
    return x**3+x**2-5*x

def miejsca_zerowe(funkcja):
    pochodna_funkcji=funkcja.diff()
    lista_mz=sympy.solve(pochodna_funkcji)
    return lista_mz

def wartosci_mz(lista_mz):
    lista_wartosci_mz=[]
    for i in range(len(lista_mz)):
        liczba=wartosc_funkcji(lista_mz[i])
        lista_wartosci_mz.append(liczba)
    return lista_wartosci_mz

a=miejsca_zerowe(funkcja)
b=wartosci_mz(a)
print(a)
print(b)

x1=np.arange(-3,2,0.1)

y1=[]
for i in x1:
    y1.append(wartosc_funkcji(i))

tytul='$f(x)=x^3+x^2-5x$'


fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x1, y1, color='m', label=funkcja)
ax.plot(a,b,'go', color='black', label="Ekstrema")
ax.axhline(0, color='black', linewidth=0.75)
ax.set_xticks([-5/3,1])
plt.title("Wykres funkcji " + tytul)
ax.legend()
plt.savefig('01_str_277094.pdf')
```
