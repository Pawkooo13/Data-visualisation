```python
import numpy as np 
import matplotlib.pyplot as plt
import random 
import sympy

x = np.linspace(0,5,100)
y = 1/(x+2)**2

X = sympy.symbols("X")
calka = sympy.integrate(1/(X+2)**2, (X, 0, 5))
calka = round(float(calka),4)
napis = r'$f(x)=\frac{1}{(x+2)^2}$'

liczba_slupkow = random.randint(1,30)

lista = []
lista2 = []

lista3 = []
lista4 = []

przedzialy = 5/liczba_slupkow


for i in range(liczba_slupkow):
    lista.append(i*przedzialy)
    
for j in lista:
    a = 1/(j+2)**2
    lista2.append(a)    

for i in range(1,liczba_slupkow+1):
    lista3.append(i*przedzialy)
    
for j in lista3:
    a = 1/(j+2)**2
    lista4.append(a)
        
    
lista_pom=lista3

for i in range(len(lista3)):
    lista3[i]=lista3[i]-przedzialy


lista5=[]
lista6=[]

for i in lista2:
    p = i*przedzialy
    lista5.append(p)

for i in lista4:
    p = i*przedzialy
    lista6.append(p) 

fig, axes = plt.subplots(1, 3, figsize=(16, 10))
axes[0].bar(lista,lista2,color = "orange", align = "edge", width = przedzialy, edgecolor = 'grey')
axes[0].plot(x, y, color = "green")
axes[0].scatter(lista,lista2,color='k',zorder=5)
axes[0].set_title("Suma = " + str(round(sum(lista5),4)))

axes[1].plot(x, y)
axes[1].fill_between(x,y,facecolor='orange', alpha=0.75, )
axes[1].annotate(napis,xy=(3,0.15),xytext=(2,0.15),color='k',size=20)
axes[1].set_title("Przedziaw jest: " + str(liczba_slupkow) + "   Pole = " + str(calka))

axes[2].bar(lista3, lista4,color = "orange", align = "edge", width = przedzialy, edgecolor = 'grey')
axes[2].plot(x,y, color = "green")
axes[2].scatter(lista_pom[1:],lista2[1:],color='k',zorder=5)
axes[2].set_title("Suma = " + str(round(sum(lista6),4)))

plt.savefig("05_str_277094.pdf")
```
