```python
from matplotlib import pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np
import sympy

napis=r'$\frac{T^2}{r^3}=const$'

x=sympy.Symbol('x')

x=np.linspace(60,11**5,1000000)

lista_T=[88.0,224.7,365.2,687.0,4331,10747,30589,59800,90560]

c=3.36*(10**18)

def wartosc_funkcji(liczba):
    c=3.36*(10**18)
    y=(liczba**2*c)**(1/3)
    return y

lista_R=[]

for i in lista_T:
    lista_R.append(wartosc_funkcji(i))

funkcja=(x**2*c)**(1/3)
    
fig=plt.figure()
ax=fig.add_subplot()
ax.set_xlim(60,11**5)
ax.plot(x,funkcja,color="green")
ax.set_yscale("log")
ax.set_xscale("log")
ax.set_xlabel("Okres orbitalny w latach [T]")
ax.set_ylabel("Półoś wielka [r]")
ax.set_facecolor('k')
ax.annotate(napis,xy=(100,10**9),xytext=(100,10**9),color='white',size=20)
plt.title("Trzecie prawo Keplera")


def getImage(path):
    return OffsetImage(plt.imread(path))

paths = ["merkury.png","wenus.png","ziemia.png","mars.png","jowisz.png","saturn.png","uran.png","neptun.png","pluton.png","sun.png"]

for i in range(len(lista_R)):
    ab = AnnotationBbox(getImage(paths[i]), (lista_T[i], lista_R[i]), frameon=False)
    ax.add_artist(ab)
    
ab=AnnotationBbox(getImage(paths[9]),(50000,12.5**7),frameon=False)
ax.add_artist(ab)

plt.savefig("03_str_277094.pdf")
```
