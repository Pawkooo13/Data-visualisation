```python
from mpl_toolkits.basemap import Basemap
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import numpy as np

fig = plt.figure(figsize=(10,8))

mapa=Basemap(llcrnrlon = -110, llcrnrlat = -40, urcrnrlon=150, urcrnrlat=60, projection='lcc', lat_0=0.01, lon_0=0.01)
mapa.drawcoastlines()

MachuPicchu = mapa(-72.32,-13.09)
Petra = mapa(35.26,30.19)
ChrystusZbawiciel = mapa(-43.12,-22.57)
TadżMahal = mapa(78.02,27.10)
ChichenItza = mapa(-88.34,20.40)
Koloseum = mapa(12.29,41.53)
MurChinski = mapa(116.57,40.43)

MachuPicchu1 = mapa(-72.32+3,-13.09)
Petra1 = mapa(35.26+3,30.19)
ChrystusZbawiciel1 = mapa(-43.12+3,-22.57)
TadżMahal1 = mapa(78.02+3,27.10)
ChichenItza1 = mapa(-88.34+3,20.40)
Koloseum1 = mapa(12.29+3,41.53)
MurChinski1 = mapa(116.57+3,40.43)


mapa.plot(MachuPicchu[0],MachuPicchu[1],'ko',color='red')
mapa.plot(Petra[0],Petra[1],'ko',color='red')
mapa.plot(ChrystusZbawiciel[0],ChrystusZbawiciel[1],'ko',color='red')
mapa.plot(TadżMahal[0],TadżMahal[1],'ko',color='red')
mapa.plot(ChichenItza[0],ChichenItza[1],'ko',color='red')
mapa.plot(Koloseum[0],Koloseum[1],'ko',color='red')
mapa.plot(MurChinski[0],MurChinski[1],'ko',color='red')
mapa.shadedrelief()

plt.title("Trasa po 7 cudach świata")

plt.annotate("Machu Picchu", xy=(MachuPicchu[0],MachuPicchu[1]),xytext=(MachuPicchu1[0],MachuPicchu1[1]),bbox=dict(boxstyle="square",facecolor='paleturquoise',pad=0.3),size=6)
plt.annotate("Petra", xy=(Petra[0],Petra[1]),xytext=(Petra1[0],Petra1[1]),bbox=dict(boxstyle="square",facecolor='paleturquoise',pad=0.3),size=6)
plt.annotate("Statua Chrystusa", xy=(ChrystusZbawiciel[0],ChrystusZbawiciel[1]),xytext=(ChrystusZbawiciel1[0],ChrystusZbawiciel1[1]),bbox=dict(boxstyle="square",facecolor='paleturquoise',pad=0.3),size=6)
plt.annotate("Tadż Mahal", xy=(TadżMahal[0],TadżMahal[1]),xytext=(TadżMahal1[0],TadżMahal1[1]),bbox=dict(boxstyle="square",facecolor='paleturquoise',pad=0.3),size=6)
plt.annotate("Chichen Itza", xy=(ChichenItza[0],ChichenItza[1]),xytext=(ChichenItza1[0],ChichenItza1[1]),bbox=dict(boxstyle="square",facecolor='paleturquoise',pad=0.3),size=6)
plt.annotate("Koloseum", xy=(Koloseum[0],Koloseum[1]),xytext=(Koloseum1[0],Koloseum1[1]),bbox=dict(boxstyle="square",facecolor='paleturquoise',pad=0.3),size=6)
plt.annotate("Mur Chiński", xy=(MurChinski[0],MurChinski[1]),xytext=(MurChinski1[0],MurChinski1[1]),bbox=dict(boxstyle="square",facecolor='paleturquoise',pad=0.3),size=6)

a,=mapa.drawgreatcircle(12.29,41.53,-88.34,20.40,color='orange')
b,=mapa.drawgreatcircle(-88.34,20.40,-72.32,-13.09,color='orange')
c,=mapa.drawgreatcircle(-72.32,-13.09,-43.12,-22.57,color='orange')
d,=mapa.drawgreatcircle(-43.12,-22.57,35.26,30.19,color='orange')
e,=mapa.drawgreatcircle(35.26,30.19,78.02,27.10,color='orange')
f,=mapa.drawgreatcircle(78.02,27.10,116.57,40.43,color='orange')
g,=mapa.drawgreatcircle(116.57,40.43,12.29,41.53,color='orange')

listax=[]
listay=[]

def funkcja(a):
    x,y = a.get_data()
    listax.append(x)
    listay.append(y)

funkcja(a)
funkcja(b)
funkcja(c)
funkcja(d)
funkcja(e)
funkcja(f)
funkcja(g)

dane_x=[]
dane_y=[]

a.remove()
b.remove()
c.remove()
d.remove()
e.remove()
f.remove()
g.remove()


for i in range(7):
    for j in range(len(listax[i])):
        dane_x.append(listax[i][j])
        dane_y.append(listay[i][j])

def animuj(i,dane_x,dane_y):
    plt.plot(dane_x[:i],dane_y[:i],color='orange')
    
ani = FuncAnimation(fig,animuj,fargs=(list(dane_x),list(dane_y)),frames=len(dane_x),interval=0.1)
ani.save('11_str_277094.gif',writer=PillowWriter(fps=25),dpi=200)
plt.show()
```
